```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_src = cv2.imread('chess_2.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.title('Перетворене зображення')
plt.axis('off')
plt.show()
```

```python
A = [216,68]
B = [472, 115]
C = [381, 347]
D = [222, 312]
pts_src = np.array([A, B, C, D])
```

```python
im_dst = cv2.imread('chess_1.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.title('Базове зображення')
plt.axis('off')
plt.show()
```

```python
A = [447,72]
B = [444, 230]
C = [214, 219]
D = [210, 79]
pts_dst = np.array([D,A,B,C])
```

```python
h, status = cv2.findHomography(pts_src, pts_dst,)
```

```python
h
```

```python
im_out = cv2.warpPerspective(im_src, h,
                             (im_dst.shape[1],
                              im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.title('Відтворене зображення')
plt.axis('off')
plt.show()
```
